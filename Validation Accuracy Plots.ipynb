```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the epoch performance comparison data
epoch_results_df = pd.read_csv('epoch_performance_comparison.csv')

# Plot validation accuracy for each model on a single plot
plt.figure(figsize=(6.5, 4.5))

# Group the dataframe by model and plot each one
for model_name, group in epoch_results_df.groupby('Model'):
    plt.plot(group['Epoch'], group['Val_Accuracy'], label=f'{model_name}')

plt.axhline(y=0.8, color='r', linestyle='--', label='80% Acc')
plt.axhline(y=0.9, color='g', linestyle='--', label='90% Acc')
plt.title('Validation Accuracy Trends for Deep Learning Models', fontsize=15)
plt.xlabel('Epochs', fontsize=13)
plt.ylabel('Validation Accuracy', fontsize=13)
plt.legend(fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the epoch performance comparison data for ResNet-50
epoch_results_df = pd.read_csv('resnet_epoch_performance_comparison.csv')

# Plot validation accuracy for ResNet-50 model
plt.figure(figsize=(6.5, 4.5))

# Group the dataframe by model and plot each one
for model_name, group in epoch_results_df.groupby('Model'):
    plt.plot(group['Epoch'], group['Val_Accuracy'], label=f'ResNet-50')

plt.axhline(y=0.8, color='r', linestyle='--', label='80% Acc')
plt.axhline(y=0.9, color='g', linestyle='--', label='90% Acc')
plt.title('Validation Accuracy Trends for ResNet-50', fontsize=15)
plt.xlabel('Epochs', fontsize=13)
plt.ylabel('Validation Accuracy', fontsize=13)
plt.legend(fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```
