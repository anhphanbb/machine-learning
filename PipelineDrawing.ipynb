```python
# Adjust the diagram to have even smaller boxes and reverse arrow directions
fig, ax = plt.subplots(figsize=(10, 6))

# Function to add an even smaller box with text and smaller height
def add_smaller_box(ax, text, xy, box_color='#D3D3D3'):
    box = patches.FancyBboxPatch(xy, width=2.5, height=0.5, boxstyle="round,pad=0.3", edgecolor='black', facecolor=box_color)
    ax.add_patch(box)
    ax.text(xy[0] + 1.25, xy[1] + 0.3, text, ha='center', va='center', fontsize=14)
    #ax.text(xy[0] + 1.25, xy[1] + 0.3, text, ha='center', va='center', fontsize=14, fontweight='bold')

# Add boxes and arrows in a counterclockwise loop with even smaller height
left_steps = [
    ("Current Model", (1, 4.5)),
    ("Predict New Data", (1, 3)),
    ("Human Verification", (1, 1.5))
]

right_steps = [
    ("Incorporate Verified Data", (4.5, 1.5)),
    ("Retrain Model", (4.5, 3)),
    ("Improved Model", (4.5, 4.5))
]

for step in left_steps:
    add_smaller_box(ax, step[0], step[1])

for step in right_steps:
    add_smaller_box(ax, step[0], step[1])

# Draw arrows in a counterclockwise loop
arrows = [
    
    
    ((2.25, 2.75), (2.25, 2.25)),
    ((2.25, 4.25), (2.25, 3.75)),
    ((3.75, 1.75), (4.25, 1.75)),
    ((5.75, 2.3), (5.75, 2.75)),
    ((5.75, 3.75), (5.75, 4.25)),
    ((4.25, 4.75), (3.75, 4.75))
]

for arrow in arrows:
    ax.annotate("", xy=arrow[1], xytext=arrow[0],
                arrowprops=dict(arrowstyle="->", color='black'))

# Set limits and remove axes
ax.set_xlim(0, 9)
ax.set_ylim(0, 6)
ax.axis('off')

# Display the diagram
plt.show()
```
