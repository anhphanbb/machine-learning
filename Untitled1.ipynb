```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Read the grid search results from the CSV file
cv_results = pd.read_csv('svm_grid_search_results.csv')

# Filter the results based on gamma value
results_scale = cv_results[cv_results['param_gamma'] == 'scale']
results_auto = cv_results[cv_results['param_gamma'] == 'auto']

# Extract unique C values and kernels for plotting
C_values = sorted(results_scale['param_C'].unique())
kernels = ['linear', 'poly', 'rbf']

# Create a 2D array for each plot
heatmap_data_scale = np.zeros((len(C_values), len(kernels)))
heatmap_data_auto = np.zeros((len(C_values), len(kernels)))

# Populate the arrays with mean test scores
for idx, row in results_scale.iterrows():
    C_index = C_values.index(row['param_C'])
    kernel_index = kernels.index(row['param_kernel'])
    heatmap_data_scale[C_index, kernel_index] = row['mean_test_score']

for idx, row in results_auto.iterrows():
    C_index = C_values.index(row['param_C'])
    kernel_index = kernels.index(row['param_kernel'])
    heatmap_data_auto[C_index, kernel_index] = row['mean_test_score']

# Define a consistent color palette for the heatmaps
heatmap_cmap = sns.light_palette("#1f77b4", as_cmap=True)

# Increase font size
plt.rcParams.update({'font.size': 16})

# Plot heatmap for gamma=scale
fig, ax = plt.subplots(1, 2, figsize=(20, 8))

sns.heatmap(heatmap_data_scale, annot=True, xticklabels=kernels, yticklabels=C_values, cmap=heatmap_cmap, cbar=False, ax=ax[0])
ax[0].set_xlabel('Kernel')
ax[0].set_ylabel('C')
ax[0].set_title('Grid Search Accuracy (gamma=scale)')

sns.heatmap(heatmap_data_auto, annot=True, xticklabels=kernels, yticklabels=C_values, cmap=heatmap_cmap, ax=ax[1])
ax[1].set_xlabel('Kernel')
ax[1].set_ylabel('C')
ax[1].set_title('Grid Search Accuracy (gamma=auto)')

# Add a single color bar for both heatmaps
cbar_ax = fig.add_axes([0.92, 0.3, 0.02, 0.4])
fig.colorbar(ax[1].collections[0], cax=cbar_ax)

plt.show()
```
