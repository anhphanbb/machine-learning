```python
import matplotlib.pyplot as plt

# Data distribution
total_images = 28690
clear_images = 21304
cloud_images = 7386

balanced_clear_images = 2663
balanced_cloud_images = 2462

# Data splitting
train_split = 0.7
validation_split = 0.2
test_split = 0.1

# Balanced dataset
balanced_total = balanced_clear_images + balanced_cloud_images

# Data splitting counts
train_count = balanced_total * train_split
validation_count = balanced_total * validation_split
test_count = balanced_total * test_split

# Function to add percentage and count in pie chart
def autopct_generator(values):
    def inner_autopct(pct):
        total = sum(values)
        val = int(round(pct * total / 100.0))
        return f'{pct:.1f}%\n({val:d})'
    return inner_autopct

# Plotting
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Distribution of original dataset
axes[0].pie([clear_images, cloud_images], labels=['Clear', 'Cloud'], autopct=autopct_generator([clear_images, cloud_images]), startangle=140)
axes[0].set_title('Distribution of Original Dataset')

# Distribution of balanced dataset
axes[1].pie([balanced_clear_images, balanced_cloud_images], labels=['Clear', 'Cloud'], autopct=autopct_generator([balanced_clear_images, balanced_cloud_images]), startangle=140)
axes[1].set_title('Distribution of Balanced Dataset')

plt.show()

# Data splitting
splits = ['Train', 'Validation', 'Test']
split_counts = [train_count, validation_count, test_count]

plt.figure(figsize=(5, 5))
plt.pie(split_counts, labels=splits, autopct=autopct_generator(split_counts), startangle=140)
plt.title('Data Splitting')
plt.show()
```
